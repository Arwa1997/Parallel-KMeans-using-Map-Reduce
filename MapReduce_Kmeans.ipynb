```
#getInitialCentroids.py
from random import choice
import pandas as pd
import numpy as np
def getInitialCentroids(filepath, k):
  data =[]
  with open(filepath) as fp:
      for line in fp:
        try:
          line = line.strip()
          cords = line.split(' ') 
          print("cords",cords)
          tuple = []
          for cord in cords:
            tuple.append(float(cord))
          data.append(tuple)
        except:
            break
      print("data",data)
      
      centroids =[]
      for i in range (0,k):
        centroid = data[choice(range(0,150))][:]
        centroids.append(centroid)
      print(centroids)
      cntrs = pd.DataFrame(centroids)
      np.savetxt(r'centroids.txt', cntrs.values, fmt='%d')

if __name__ == '__main__':
  #using iris dataset
  getInitialCentroids("iris.txt",3)

```

```
#mapper.py
import sys
from math import sqrt

# get initial centroids from a txt file and add them in an array

def getCentroids(filepath):
    centroids = []
    
    with open(filepath) as fp:
        line = fp.readline()
        while line:
            if line:
                try:
                  line = line.strip()
                  cord = line.split()
                  for i in range(len(cord)):
                    cord[i]=float(cord[i])
                  centroids.append(cord)
                except:
                    break
            else:
                break
            line = fp.readline()

    fp.close()
    #print("centroids", centroids)
    return centroids

# create clusters based on initial centroids
def createClusters(centroids):
    # 
    #for line in sys.stdin:
    fw = open("output.txt","w+")
    with open('iris.txt') as fp:
      line = fp.readline()
      while line:
        line = line.strip()
        cord = line.split(' ') 
        min_dist = 100000000000000 #infinity
        index = -1
        for centroid in centroids:
            try:
                for i in range(len(cord)):
                  cord[i]=float(cord[i])
            except ValueError:
                continue

            # euclidian distance from every point of dataset to centroids
            total_dist = 0 
            for i in range(len(cord)):
              total_dist = total_dist + pow(cord[i] - centroid[i], 2)
            cur_dist = sqrt(total_dist)

            # find closest centroid
            if cur_dist <= min_dist:
                min_dist = cur_dist
                index = centroids.index(centroid)

        var = "%s\t" % ( index )
        for i in cord:
          var = var + str(i) + " "
        print(var)
        line = fp.readline()
        fw.write(var+"\n")
              
      fw.close()

        
if __name__ == "__main__":
    centroids = getCentroids('centroids.txt')
    createClusters(centroids)
```
